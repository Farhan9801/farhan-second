```python
class LL:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def inputL():
    list=[int(i) for i in input().split()]
    head=None
    for current in list:
        if current==-1:
            break
        
        newnode=LL(current)
        if head==None:
            head=newnode
            
        else:
            curr=head
            while curr.next is not None:
                curr=curr.next
                
            curr.next=newnode
    return head
        
def printL(head):
    while head is not None:
        print(str(head.data)+'->',end=" ")
        head=head.next
    
    print('None')
    
def lenL(head):
    l=0
    while head is not None:
        l+=1
        head=head.next
        
    return l

def insertL(head,i,data):
    if i<0 and i>len(head):
        return head
    j=0
    prev=None
    curr=head
    while j<i:
        prev=curr
        curr=curr.next
        j+=1
    newnode=LL(data)
    if prev is not None:
        prev.next=newnode
    else:
        head=newnode
    newnode.next=curr
    
    return head

def deleteL(head,i):
    j=0
    prev=None
    curr=head
    while j<i:
        prev=curr
        curr=curr.next
        j+=1
        if j==i:
            if curr is not None:
                prev.next=curr.next
            else:
                prev.next=None
    if prev==None:
        head=head.next
    
    return head

def reverseLL(head):
    if head is None or head.next is None:
        return head,head
    smallhead,smalltail=reverseLL(head.next)
    smalltail.next=head
    head.next=None
    return smallhead,head

def reverseLL2(head):
    if head is None or head.next is None:
        return head
    smallhead=reverseLL2(head.next)
    tail=head.next
    tail.next=head
    head.next=None
    return smallhead

def reverseLLi(head):
    curr=head
    prev=None
    while curr is not None:
                
        nextt=curr.next
        curr.next=prev
        prev=curr
        curr=nextt
    return prev

def midLL(head):
    x=lenL(head)//2
    if x%2!=0:
        for i in range(x):
            head= head.next
    else:
        for i in range(x-1):
            head=head.next
    return head.data

def midLL2(head):
    slow=head
    fast=head
    while fast.next!=None and fast.next.next!=None:
        slow=slow.next
        fast=fast.next.next
    return slow.data

def mergeLL(h1,h2):
    head=None
    tail=None
    
    if h1.data<h2.data:
        head=h1
        tail=h1
        h1=h1.next
        
    else:
        head=h2
        tail=h2
        h2=h2.next
    while h1!=None and h2!=None:
        if h1.data<h2.data:
            tail.next=h1
            h1=h1.next
            tail=tail.next
        
        else:
            tail.next=h2
            h2=h2.next
            tail=tail.next
    if h1==None:
        while h2 is not None:
            tail.next=h2
            h2=h2.next
            tail=tail.next
    else:
        while h1 is not None:
            tail.next=h1
            h1=h1.next
            tail=tail.next
            
    return head
    
    
    
    
    
        
h1=inputL()
h2=inputL()
print(h2.data)
printL(h2)
print(h1.data)
head=mergeLL(h1,h2)
printL(head)
```
