```python
class BinaryNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
```

```python
def BTinput():
    rootdata=int(input())
    if rootdata==-1:
        return None
    root=BinaryNode(rootdata)
    root.left=BTinput()
    root.right=BTinput()
    return root
```

```python
def inputBST(lst):
    if not lst:
        return None
    x=len(lst)//2
    root=BinaryNode(lst[x])
    leftnode=inputBST(lst[:x])
    rightnode=inputBST(lst[x+1:])
    root.left=leftnode
    root.right=rightnode
    return root
```

```python
def maxNode(root):
    if root==None:
        return -100000000
    leftmax=maxNode(root.left)
    rightmax=maxNode(root.right)
    return max(leftmax,rightmax,root.data)
```

```python
def minNode(root):
    if root==None:
        return 100000000
    leftmin=minNode(root.left)
    rightmin=minNode(root.right)
    return min(leftmin,rightmin,root.data)
```
